```python
# GRAPH 
import matplotlib.pyplot as plt

# x axis values
x = [1,3,5,6,7,9,10]

#  y axis values
y = [1,2,4,5,6,7,8]

# plotting the points
plt.plot(x, y, color='green', linestyle='solid', linewidth = 5, marker='o', markerfacecolor='red', markersize = 10)

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# title
plt.title('dotted line graph')

#  show the plot
plt.show()
```
